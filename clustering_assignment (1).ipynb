```python
# Clustering Assignment - Full Notebook


# ============================================================
# THEORY QUESTIONS (WITH ANSWERS)
# ============================================================

# 1. What is unsupervised learning?
"""
Unsupervised learning is a type of machine learning where the model learns patterns
and structures from unlabeled dataâ€”meaning no predefined output is provided.
"""

# 2. How does K-Means clustering work?
"""
K-Means partitions data into k clusters by:
1. Choosing k initial centroids
2. Assigning each point to the nearest centroid
3. Updating centroids as mean of assigned points
4. Repeating steps 2-3 until convergence.
"""

# 3. Explain dendrogram in hierarchical clustering
"""
A dendrogram is a tree-like diagram that shows how clusters merge at each step
of hierarchical clustering.
"""

# 4. Difference between K-Means and Hierarchical Clustering
"""
K-Means requires predefined number of clusters and is centroid-based.
Hierarchical clustering builds a tree of clusters and does not need predefined k.
"""

# 5. Advantages of DBSCAN over K-Means
"""
- No need to specify number of clusters
- Can find arbitrarily shaped clusters
- Handles noise/outliers well
"""

# 6. When to use Silhouette Score
"""
Use it to evaluate how well clusters are separated and how compact they are.
"""

# 7. Limitations of Hierarchical Clustering
"""
- Computationally expensive
- Sensitive to noise
- No ability to correct wrong merges
"""

# 8. Why feature scaling is important in K-Means
"""
Because K-Means uses Euclidean distance; unscaled features distort cluster formation.
"""

# 9. How DBSCAN identifies noise
"""
Points not belonging to any dense region (core or border) are labeled as -1.
"""

# 10. Define inertia in K-Means
"""
Sum of squared distances between each point and its assigned centroid.
"""

# 11. Elbow Method
"""
Technique to find optimal k by plotting inertia vs k and selecting elbow point.
"""

# 12. Density in DBSCAN
"""
Density = number of points within a radius (eps). Used to detect dense regions.
"""

# 13. Can hierarchical clustering work on categorical data?
"""
Yes, but requires special distance metrics (e.g., Hamming distance).
"""

# 14. Negative Silhouette Score
"""
Indicates samples are misclassified (assigned to wrong cluster).
"""

# 15. Linkage criteria
"""
Rules to decide distance between clusters (single, complete, average, ward).
"""

# 16. Why K-Means fails on varying sizes/densities
"""
Because centroid-based clustering assumes spherical and equally sized clusters.
"""

# 17. Core parameters of DBSCAN
"""
- eps: neighborhood radius
- min_samples: minimum points to form dense region
"""

# 18. K-Means++ improvement
"""
Chooses initial centroids far apart, leading to better convergence.
"""

# 19. What is agglomerative clustering?
"""
Bottom-up hierarchical technique merging closest clusters step by step.
"""

# 20. Why Silhouette Score is better than inertia
"""
Because it considers both compactness and separation; inertia only measures compactness.
"""


# ============================================================
# PRACTICAL / CODING QUESTIONS
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles, load_iris, load_wine, load_breast_cancer, load_digits
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE

# 1. Generate synthetic data with 4 centers and apply K-Means
X, y = make_blobs(n_samples=500, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("K-Means with 4 Blobs")
plt.show()

# 2. Load Iris dataset and use Agglomerative Clustering
iris = load_iris()
X = iris.data
agg = AgglomerativeClustering(n_clusters=3)
pred = agg.fit_predict(X)
print("First 10 predicted labels:", pred[:10])

# 3. Make moons + DBSCAN
X, y = make_moons(n_samples=400, noise=0.1)
db = DBSCAN(eps=0.2, min_samples=5)
labels = db.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("DBSCAN on Moons")
plt.show()

# 4. Wine dataset + K-Means after standardizing
wine = load_wine()
X = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X)
unique, counts = np.unique(labels, return_counts=True)
print("Cluster sizes:", dict(zip(unique, counts)))

# 5. make_circles + DBSCAN
X, y = make_circles(n_samples=500, noise=0.05, factor=0.5)
db = DBSCAN(eps=0.15)
labels = db.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("DBSCAN on Circles")
plt.show()

# ... Continue adding all practical questions similarly

# Continue Practical Questions

# 6. Breast Cancer dataset + MinMaxScaler + KMeans
bc = load_breast_cancer()
X = MinMaxScaler().fit_transform(bc.data)
kmeans = KMeans(n_clusters=2)
labels = kmeans.fit_predict(X)
print("Cluster Centroids:\n", kmeans.cluster_centers_)

# 7. Blobs with varying std + DBSCAN
X, y = make_blobs(n_samples=600, centers=3, cluster_std=[0.3, 1.2, 2.5])
labels = DBSCAN(eps=0.5).fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("DBSCAN on varying std blobs")
plt.show()

# 8. Digits dataset + PCA 2D + KMeans
X = load_digits().data
X_pca = PCA(n_components=2).fit_transform(X)
labels = KMeans(n_clusters=10).fit_predict(X_pca)
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels)
plt.title("Digits KMeans PCA 2D")
plt.show()

# 9. Silhouette scores for k=2 to 5
X, _ = make_blobs(n_samples=500, centers=4)
scores = []
ks = [2,3,4,5]
for k in ks:
    labels = KMeans(n_clusters=k).fit_predict(X)
    scores.append(silhouette_score(X, labels))
plt.bar(ks, scores)
plt.title("Silhouette Scores")
plt.show()

# 10. Iris + hierarchical dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage
X = load_iris().data
Z = linkage(X, method='average')
plt.figure(figsize=(10,6))
dendrogram(Z)
plt.title("Dendrogram - Iris")
plt.show()

# 11. Overlapping blobs + KMeans + decision boundaries
X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.0)
kmeans = KMeans(n_clusters=3).fit(X)
labels = kmeans.labels_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Overlapping KMeans")
plt.show()

# 12. Digits dataset + t-SNE + DBSCAN
X = load_digits().data
X_tsne = TSNE(n_components=2, perplexity=30).fit_transform(X)
labels = DBSCAN(eps=2.5).fit_predict(X_tsne)
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=labels)
plt.title("Digits t-SNE DBSCAN")
plt.show()

# 13. make_blobs + Agglomerative complete linkage
X, y = make_blobs(n_samples=400, centers=4)
agg = AgglomerativeClustering(linkage='complete', n_clusters=4)
labels = agg.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Agglomerative Complete")
plt.show()

# 14. Breast Cancer inertia for k=2 to 6
X = load_breast_cancer().data
X = StandardScaler().fit_transform(X)
inertias = []
ks = [2,3,4,5,6]
for k in ks:
    inertias.append(KMeans(n_clusters=k).fit(X).inertia_)
plt.plot(ks, inertias)
plt.title("Inertia values")
plt.show()

# 15. Concentric circles + Agglomerative single linkage
X, y = make_circles(n_samples=500, noise=0.02)
labels = AgglomerativeClustering(linkage='single', n_clusters=2).fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Single Linkage Circles")
plt.show()

# 16. Wine + DBSCAN after scaling
wine = load_wine()
X = StandardScaler().fit_transform(wine.data)
labels = DBSCAN(eps=1.2).fit_predict(X)
clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Clusters (excluding noise):", clusters)

# 17. make_blobs + KMeans + plot centers
X, y = make_blobs(n_samples=500, centers=4)
kmeans = KMeans(n_clusters=4).fit(X)
labels = kmeans.labels_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='x')
plt.title("KMeans with Centers")
plt.show()

# 18. Iris + DBSCAN noise count
X = load_iris().data
labels = DBSCAN(eps=0.6).fit_predict(X)
print("Noise samples:", list(labels).count(-1))

# 19. Moons + KMeans visualization
X, y = make_moons(n_samples=400, noise=0.1)
labels = KMeans(n_clusters=2).fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("KMeans on Moons")
plt.show()

# 20. Digits 3D PCA + KMeans
X = load_digits().data
X_pca = PCA(n_components=3).fit_transform(X)
labels = KMeans(n_clusters=10).fit_predict(X_pca)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=labels)
plt.title("Digits 3D PCA KMeans")
plt.show()

# END OF NOTEBOOK
```
